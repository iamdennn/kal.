```python
import numpy as np

def gaussian_elimination(A, b):
    # Gabungkan A dan b menjadi matriks augmentasi
    A = A.astype(float)
    b = b.astype(float).reshape(-1, 1)
    n = len(b)
    matrix = np.hstack([A, b])
    
    print("Matriks Augmentasi Awal:")
    print(matrix)
    print("-" * 30)

    # Proses Eliminasi (Forward Elimination)
    for i in range(n):
        # Cari pivot terbesar untuk stabilitas (Opsional tapi disarankan)
        pivot = i + np.argmax(np.abs(matrix[i:, i]))
        matrix[[i, pivot]] = matrix[[pivot, i]]
        
        for j in range(i + 1, n):
            faktor = matrix[j, i] / matrix[i, i]
            matrix[j, i:] -= faktor * matrix[i, i:]
            
    print("Matriks Setelah Eliminasi (Eselon Baris):")
    print(np.round(matrix, 2))
    print("-" * 30)

    # Proses Substitusi Balik (Back Substitution)
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (matrix[i, -1] - np.dot(matrix[i, i+1:n], x[i+1:n])) / matrix[i, i]
        
    return x

# --- PENYELESAIAN ---

# 1. SPL Tiga Variabel (3x3)
A3 = np.array([[1, 2, 1], 
               [1, 3, 2], 
               [2, 1, 2]])
b3 = np.array([6, 9, 6])

print("=== PENYELESAIAN SPL 3 VARIABEL ===")
solusi3 = gaussian_elimination(A3, b3)
print(f"Solusi [x, y, z]: {solusi3}\n\n")
```

```python
# 2. SPL Empat Variabel (4x4)
A4 = np.array([[1, 1, 1, 1],
               [2, 1, -1, 1],
               [1, -2, 2, 3],
               [1, -1, -1, 2]])
b4 = np.array([10, 5, 10, 3])

print("=== PENYELESAIAN SPL 4 VARIABEL ===")
solusi4 = gaussian_elimination(A4, b4)
print(f"Solusi [w, x, y, z]: {solusi4}")
```
